```python
import pandas as pd
from datasets import Dataset
from PIL import Image
import os

#
image_paths = [
    "/content/Cat.jpg",
    "/content/dog.jpg",
    "/content/tiger.jpg",
    "/content/elephant.jpg",
    "/content/lion.jpg",
    "/content/horse.jpg",
    "/content/zebra.jpg"
]
descriptions = [
    "cat",
    "dog",
    "tiger",
    "elephant",
    "lion",
    "horse",
    "zebra"
]


data = {"image_path": image_paths, "text": descriptions}
df = pd.DataFrame(data)

# Create a custom dataset
dataset = Dataset.from_pandas(df)

# to load the images as PIL objects
def load_image(example):
    # Open the image from the file path and convert to RGB
    image = Image.open(example["image_path"]).convert("RGB")
    example["image"] = image  # Store the loaded image in the dataset
    return example

# Apply the function to the dataset to create the 'image' column
dataset = dataset.map(load_image)

dataset = dataset.remove_columns("image_path")

dataset = dataset.select_columns(["image", "text"])

print(dataset)
```
